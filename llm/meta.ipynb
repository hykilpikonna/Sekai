```python
import json
from pathlib import Path
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

# Configure matplotlib to use font LXGW WenKai
plt.rcParams['font.sans-serif'] = ['LXGW WenKai']
plt.rcParams['axes.unicode_minus'] = False

all_chat = json.loads(Path('data/all_chat.json').read_text('utf-8'))
print(f"Total chat messages: {len(all_chat)}")
# Filter out numerical speakers

# Convert to DataFrame
all_chat_df = pd.DataFrame(all_chat)
all_chat_df.describe()
```

```python
all_chat_df.head()
```

```python
# Print unique speakers
all_chat_df['speaker'].unique()
```

```python
import re

# Print unique faces
faces = list(all_chat_df['face'].unique())
print(f"There are {len(faces)} unique faces.")
# Remove _\d+ and count unique faces
faces = [face.replace('face_', '') for face in faces if isinstance(face, str)]
faces = set([re.sub(r'_\d+', '', face) for face in faces])
print(f"There are {len(faces)} unique faces.")
```

```python
anims = list(all_chat_df['animation'].unique())
print(f"There are {len(anims)} unique anims.")
# Print top 50 animations and their counts
anim_freq = all_chat_df['animation'].value_counts()
print(anim_freq[:50])
```

```python
# Speaker frequency of top 50 speakers
speaker_freq = all_chat_df['speaker'].value_counts()
plt.figure(figsize=(10, 10))

n = 30
sns.barplot(x=speaker_freq[:n].values, y=speaker_freq[:n].index, palette='viridis', hue=speaker_freq[:n].index)
plt.title('Top 50 Speaker Frequency')
plt.xlabel('Frequency')
plt.ylabel('Speaker')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the number of letters sent by each speaker
all_chat_df['letter_count'] = all_chat_df['text'].apply(len)
speaker_letter_counts = all_chat_df.groupby('speaker')['letter_count'].sum()

# Get the top 30 speakers by the number of letters sent
top_speakers = speaker_letter_counts.nlargest(30)

# Plot the results
plt.figure(figsize=(10, 10))
sns.barplot(x=top_speakers.values, y=top_speakers.index, palette='viridis', hue=top_speakers.index)
plt.title('Top 30 Speakers by Number of Letters Sent')
plt.xlabel('Number of Letters')
plt.ylabel('Speaker')
plt.legend(title='Speaker')
plt.show()
```
